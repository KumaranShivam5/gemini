```python
import numpy as np 
import pandas as pd 
from matplotlib import pyplot as plt 
from astropy.io import fits
```

```python
data = fits.open('../data/binaries/LMXB_NS_v2.fits')[1].data
data = pd.DataFrame(data)
data = data[['A_NAME' ,'A_RA' ,'A_DEC' ,'A_DATABASE_TABLE' ,'A_OBSERVATORY' , 'B_NAME' ,'B_RA' , 'B_DEC' ,'XOFFSET_']]
display(data)
display(data.describe())
data.to_csv('source_list/NS_data_flagged_v2.csv')
```
