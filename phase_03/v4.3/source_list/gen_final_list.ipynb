```python
import numpy as np 
import pandas as pd 
```

```python
src_class = 'NS'
frac = 0.9 
```

```python
data = pd.read_csv(src_class+'_data_clean_source_list.csv')
data =  data.sample(frac=1)
display(data.describe())
frac_no = int(len(data)*frac)
train_data = data.iloc[:frac_no]
test_data = data.iloc[frac_no:]
display(train_data)
display(test_data)
train_data.to_csv('final_list/'+src_class+'_train.csv')
test_data.to_csv('final_list/'+src_class+'_test.csv')
```
