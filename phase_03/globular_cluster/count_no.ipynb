```python
import pandas as pd 
from matplotlib import pyplot as plt 
import numpy as np 
```

```python
files = pd.read_csv('cl_files.txt' , names=['fname'])
files
```

```python
df = pd.DataFrame()
src_counts = []
obs_counts= []
index = 0 
for index in range(len(files)):
    df = pd.read_csv('cluster_obs_data/'+files['fname'][index], comment='#' , delimiter='\t' , low_memory=False) 
    src_counts.append(len(np.unique(df['name'])))
    obs_counts.append(len(df))
```

```python
df
```

```python
deets = files.copy()
deets.insert(1 , 'src' , src_counts)
deets.insert(2 , 'obs' , obs_counts)
deets
```
