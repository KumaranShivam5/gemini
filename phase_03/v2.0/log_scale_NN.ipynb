```python
filters = ['B' , 'U' , "M" , "S" , 'W']
```

```python
flux_list = ["FLUX_AP", "FLUX_AP90" , "FLUX_AP90_AVG" , "FLUX_AP_AVG" , "FLUX_BB_AP" , "FLUX_BB_AP90" ,"FLUX_BREMS_AP" , "FLUX_BREMS_AP90", "FLUX_PLAW_AP", ]
```

```python
display(flux_list)
```

```python
NS =  fits_to_csv('../data/NS_binary.fits' , 'rows/rows_filter_level_0.json' , 0 )
NS =  NS.dropna(axis=0 , how='all')
NS =  NS[NS['B_SIGNIFICANCE']>1]
#display(NS)
ns_clean = pd.DataFrame()
i=0
for flux in flux_list:
    flux_ap_avg_list = []
    for f in filters:
        flux_ap_avg_list.append('B_'+f+'_'+flux)
    #print(flux_ap_avg_list)
    ns_flux_ap_avg = NS[flux_ap_avg_list].replace(0,np.NaN)
    temp_mean = ns_flux_ap_avg.mean(axis=1)
    ns_flux_ap_avg_mean = ns_clean.insert(i,flux ,temp_mean)
    i+=1
display(ns_clean)
temp = ns_clean.mean(axis=1)
temp = np.log10(temp)
temp_mean = np.mean(temp)
temp_var =  np.var(temp)
temp = (temp-temp_mean)/temp_var
ns = temp.to_frame(name='NORM_LOG_MEAN_FLUX')
display(ns)
```
