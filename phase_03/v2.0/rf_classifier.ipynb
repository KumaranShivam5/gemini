```python
from sklearn.ensemble import RandomForestClassifier
import pandas as pd 
import numpy as np 
from matplotlib import pyplot as plt 
```

```python
from gen_training_data import describe_data , split_data
data =  pd.read_csv('BH_NS_compiled.csv')
#print(data)
describe_data(data)
x = data.drop([ 'index_compiled' , 'class'] , axis=1)

y = data['class']
```

```python


(x_train,y_train),(x_test,y_test) = split_data(x, y, 0.7)
_ , (x_test_copy , y_test_copy) = split_data(x, y, 0.7)
print(x_train.shape ,y_train.shape)
print(x_test.shape ,y_test.shape)

(x_test_copy , y_test_copy) , (a,b) = split_data(x, y, 0.7)
np.shape(x_test_copy)
```

```python
clf = RandomForestClassifier(n_estimators=1000 , verbose=5)
clf.fit(x_train, y_train)
```

```python
imp = clf.feature_importances_
print(imp.dtype)
```

```python
plt.plot(imp)
plt.xlabel('feature index')
plt.ylabel('relative importance')
```

```python
f_names = [c for c in x]
feat_dict = [(n,float(f)) for f,n in sorted(zip(imp , f_names), reverse=1)]
#for f in feat_dict[:50]:
#    print(f[0]+': {:.4f}'.format(f[1]))

top_f_names = [f[0] for f in feat_dict[:]]
print(top_f_names)
np.savetxt('rows/top_100_features.csv',top_f_names , fmt='%s')
```
