```python
import numpy as np 
from matplotlib import pyplot as plt 
import pandas as pd 
#display(df)
import astropy.coordinates as coord

import astropy.units as u
from astropy.coordinates import SkyCoord
```

```python
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111, projection="mollweide" ,)
source_type= ['PULSAR' , 'CV' , 'NS' , 'BH']
col = ['k' , 'blue' , 'crimson' , 'green']
for s,c in zip(source_type , col):
    df = pd.read_csv('../source_list/'+s+'_data_clean_source_list.csv')
    c_icrs = SkyCoord(ra=df['B_RA']*u.degree, dec=df['B_DEC']*u.degree, frame='icrs').galactic 
    ra = c_icrs.l
    ra = ra.wrap_at(180*u.degree)
    dec = c_icrs.b
    ax.scatter(ra.radian, dec.radian, s=10 , label=s , color=c)

c_icrs = SkyCoord(ra=305.895333*u.degree, dec=-44.889114*u.degree, frame='icrs')
ax.scatter(c_icrs.ra.wrap_at(180*u.degree).radian , c_icrs.dec.radian , s = 100 , marker='o')
plt.grid()
plt.legend()
plt.show()
```
