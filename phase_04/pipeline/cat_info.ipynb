```python
import numpy as np 
import pandas as pd 
from matplotlib import pyplot as plt 
from astropy.io import fits
from IPython.display import display
```

```python
data = fits.open('../data/pulsars/atnf.fits')[1].data
data = pd.DataFrame(data)
data = data[['A_NAME' ,'A_RA' ,'A_DEC' , 'B_NAME' ,'B_RA' , 'B_DEC' ,'XOFFSET_']]
display(data)
#display(data.describe())
data_at = data.copy().set_index('A_NAME')
data_at = data_at.rename(columns={'A_RA':'RA' , 'A_DEC' :'DEC' , 'A_NAME' : 'NAME'})
data.to_csv('../source_list/pulsar_atnf.csv')
```
